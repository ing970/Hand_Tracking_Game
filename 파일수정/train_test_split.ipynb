```python
import os
import numpy as np

# set random seed at 42
np.random.seed(42)

# BASIC CONFIG (필요한 파라미터 값 여기서 설정)

data_dir = '../data/'
train_dir = data_dir + 'train/'
val_dir = data_dir + 'val/'
test_dir = data_dir + 'test/'

# get all images inside first directory in train_dir
# train_img_dir = train_dir + 'images'
# train_label_dir = train_dir + 'labels'
img_list = os.listdir(train_dir)
# label_list = os.listdir(train_label_dir)
```

```python


# randomly split items in filenames to 8:2 ratio.
np.random.shuffle(img_list)

train_img_names = img_list[:int(len(img_list)*0.8)]
val_img_names = img_list[int(len(img_list)*0.8):]

print('TOTAL IMAGES: ', len(img_list))
print('TRAIN IMAGES: ', len(train_img_names))
print('VAL IMAGES: ', len(val_img_names))

# move val_img_names to val_dir + 'images'
for img_name in val_img_names:
  os.rename(train_dir + img_name, val_dir + img_name)

```

```python
val_dir = '/content/drive/MyDrive/00DataScience/ds_study/DLproject/yolo_block/data/merge_data/val/'

val_img_dir = os.path.join(val_dir, 'images')
val_label_dir = os.path.join(val_dir, 'labels')

val_img_list = os.listdir(val_img_dir)

val_label_names = []

for img_name in val_img_list:
    # Splitting by the last dot to get the correct file name
    base_name, extension = img_name.rsplit('.', 1)
    val_label_names.append(base_name + '.txt')

# train dir 들어가서 라벨 옮기기
for label_name in val_label_names:
    os.rename(train_label_dir + '/' + label_name, val_dir + 'labels/' + label_name)
```
